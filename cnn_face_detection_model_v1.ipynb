```python
import dlib
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline 
```

```python
#构建模型，导入dat文件
face_detector = dlib.cnn_face_detection_model_v1("/mnt/hgfs/Ubuntu/dlib_model/dat/mmod_human_face_detector.dat")
```

```python
img = cv2.imread("/mnt/hgfs/Ubuntu/dlib_model/images/bigbang.jpg")
dets = face_detector(img)
'''
dets是一个mmod_rectangles对象。   
这个对象包含有2个成员变量：   
dlib.rectangle类，表示对象的位置；上下左右   
dlib.confidence，表示置信度。  
'''
```

```python

for i, d in enumerate(dets):
    face = d.rect
    print("Detection {}: Left: {} Top: {} Right: {} Bottom: {} Confidence: {}".format(i, face.left(), face.top(), face.right(), d.rect.bottom(), d.confidence))
    left = face.left()
    top = face.top()
    right = face.right()
    bottom = face.bottom()
    if d.confidence > 1:
        cv2.rectangle(img, (left, top), (right, bottom), (0, 255, 0), 3)

plt.figure(figsize=(20, 8))
plt.imshow(img)
plt.axis("off"), plt.show()
```
