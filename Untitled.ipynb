```python
import dlib
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline 
```

```python
detector = dlib.get_frontal_face_detector() #获取人脸分类器
predictor = dlib.shape_predictor("/mnt/hgfs/Ubuntu/dlib_model/dat/shape_predictor_68_face_landmarks.dat")    # 获取人脸检测器
```

```python
img = cv2.imread("/mnt/hgfs/Ubuntu/dlib_model/images/bigbang.jpg")
dets = detector(img) # 返回dlib.rectangle对象
```

```python
for i, face in enumerate(dets):
    print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(i, face.left(), face.top(), face.right(), face.bottom()))
    left = face.left()
    top = face.top()
    right = face.right()
    bottom = face.bottom()
    cv2.rectangle(img, (left, top), (right, bottom), (0, 255, 0), 3)
    shape = predictor(img, face)  # 寻找人脸的68个标定点 
    # print(shape)
    # print(shape.num_parts)
    # 遍历所有点，打印出其坐标，并用蓝色的圈表示出来
    for index, pt in enumerate(shape.parts()):
        # print('Part {}: {}'.format(index, pt))
        pt_pos = (pt.x, pt.y)
        cv2.circle(img, pt_pos, 2, (255, 0, 0), 1)

plt.figure(figsize=(20, 8))
plt.imshow(img)
plt.axis("off"), plt.show()
```
